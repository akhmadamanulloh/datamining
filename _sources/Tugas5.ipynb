````python
# membuat dataframe dengan data training tanpa label kelas
X = data.drop(columns=["variety"])#Mengecek apakah label kelas sudah terhapus
X.head()
````

````python
# persentase jumlah data test
persentase_test_data = 0.3
````

````python
# target terpisah

# nilai
X=data.iloc[:,0:4].values

# kelas
y=data.iloc[:,4].values
````

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = persentase_test_data, random_state=0)
````

````python
# penggolongan
gaussian = GaussianNB()
# 
gaussian.fit(X_train, y_train)
Y_pred = gaussian.predict(X_test) 
accuracy_nb=round(accuracy_score(y_test,Y_pred)* 100, 2)
acc_gaussian = round(gaussian.score(X_train, y_train) * 100, 2)

cm = confusion_matrix(y_test, Y_pred)
accuracy = accuracy_score(y_test,Y_pred)
precision =precision_score(y_test, Y_pred,average='micro')
recall =  recall_score(y_test, Y_pred,average='micro')
f1 = f1_score(y_test,Y_pred,average='micro')
print('Konfusi Matrix\n',cm)
print('Akurasi_Naive Bayes: %.3f' %accuracy)
print('Ketepatan_Naive Bayes: %.3f' %precision)
print('Recall_Naive Bayes: %.3f' %recall)
print('f1-score_random_Forest : %.3f' %f1)
````
